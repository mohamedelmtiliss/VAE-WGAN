```python
import requests
import os
from pathlib import Path
```

```python
base_url = "https://firms.modaps.eosdis.nasa.gov/data/country/modis/{year}/modis_{year}_Morocco.csv"
```

```python
output_dir = Path("morocco_modis_data")
output_dir.mkdir(exist_ok=True)
```

```python
# Download data from 2000 to 2024
for year in range(2000, 2025):
    url = base_url.format(year=year)
    filename = output_dir / f"modis_{year}_Morocco.csv"
    
    print(f"Downloading data for {year}...")
    
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an error for bad status codes
        
        # Save the file
        with open(filename, 'wb') as f:
            f.write(response.content)
        
        print(f"✓ Successfully downloaded: {filename}")
        
    except requests.exceptions.RequestException as e:
        print(f"✗ Failed to download {year}: {e}")
print("\nDownload complete!")
```
